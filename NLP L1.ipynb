```python
import io
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from pdfminer.pdfpage import PDFPage
```

```python
def convert_pdf_to_txt(path):
    rsrcmgr = PDFResourceManager()
    retstr = io.StringIO()
    #codec = ''
    laparams = LAParams()
    device = TextConverter(rsrcmgr, retstr, laparams=laparams)
    fp = open(path, 'rb')
    interpreter = PDFPageInterpreter(rsrcmgr, device)
    caching = True
    pagenos = set()
    maxpages=1
    password=''
    for page in PDFPage.get_pages(fp, pagenos, maxpages= maxpages, password=password, caching=caching, check_extractable=True):
        interpreter.process_page(page)
    
    text= retstr.getvalue()
    
    fp.close()
    device.close()
    retstr.close()
    return text
    
```

```python
u = convert_pdf_to_txt('Slide-7.pdf')
```

```python
print(u)
```

Extracting email IDs

```python
doc = "For more details please mail us at: xyz@abc.com, uvlakshan@gmail.com"
```

```python

addresses =re.findall(r'[\w\.-]+@[\w\.-]+', doc)
```

```python
for address in addresses:
    print(address)
```

Replacing email Ids

```python
doc = "For more details please mail us at xyz@abc.com"
```

```python
new_email_address = re.sub(r'([\w\.-]+)@([\w\.-]+)',r'pqr@mno.com', doc)
print(new_email_address)
```
